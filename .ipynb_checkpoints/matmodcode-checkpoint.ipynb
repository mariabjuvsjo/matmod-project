```python
# function for cleaning up the dataframe
# removing unwanted columns, removing rows which dont exist in all 3 df
# keep only data from month 5, get the last 40 data from each df
def clear_smhi_data_pandas(df, name, month):
    # Remove unused data columns
    df.drop(
        [
            "Från Datum Tid (UTC)",
            "Tidsutsnitt:",
            "Kvalitet",
            "Till Datum Tid (UTC)",
            "Unnamed: 5"
        ], inplace=True,
        axis=1,
       
    )

    df_month = df[df["Representativ månad"].str.endswith(month)]

    #df_month = df_month[~df_month["Representativ månad"].str.startswith(prefixes)]
    df_month = df_month.iloc[-70:]
    #df_month = df_month.set_index("Representativ månad")
    
    return df_month
```

## 5 Transformerad data

```python
df_lund_copy = df_lund_copy.reset_index()
df_lund_copy['time'] = range(1, len(df_lund_copy) + 1)
df_lund_copy["total_log"] = np.log(df_lund_copy.total)
df_lund_copy["time_log"] = np.log(df_lund_copy.time)
```

```python
propotion_train = 0.5
# träning och test data 
train = df_lund_copy.sample(frac=propotion_train, random_state=1)
test = df_lund_copy.drop(train.index)

# data för x log
train['time_log'] = np.log(train['time'])
test['time_log'] = np.log(test['time'])

# data för y log
train['total_log'] = np.log(train['total'])
test['total_log'] = np.log(test['total'])

# Kod för en linjär regression av orginal data
model = LinearRegression().fit(train[['time']], train['total'])

model = LinearRegression()

model.fit(train[['time']], train['total'])

# Prediktera med test data
pred = model.predict(test[['time']])
```

```python
# Kod för en linjär regression av log x axel
log_model = LinearRegression()
log_model.fit(train[['time_log']], train['total'])

# Prediktera med test data
pred_log = log_model.predict(test[['time_log']])
x_log = np.linspace(0, 5, 100)
draw_exp_model = log_model.predict(x_log.reshape(-1, 1))
```

```python

# Kod för en linjär regression av log y axel
log_y_model = LinearRegression()
log_y_model.fit(train[['time']], train['total_log'])

# Prediktera med test data
pred_log_y = log_y_model.predict(test[['time']])
# pred_log_y = np.exp(pred_log_y)

# pred_log_y = np.exp(pred_log_y)
```

```python
# Kod för en linjär regression av log y och x axel
log_log_model = LinearRegression()
log_log_model.fit(train[['time_log']], train['total_log'])

# Prediktera med test data
pred_log_log = log_log_model.predict(test[['time_log']])
pred_log_log_back = np.exp(pred_log_log)
```

```python
# räkna ut de olika MSE
mse = np.mean((pred - test['total'])**2)

mse_log_x = np.mean((pred_log - test['total'])**2)



# Beräkna residualer
residual_log_y = test['total'] - np.exp(pred_log_y)

# Skriv ut MSE
mse_exp_y = np.mean(residual_log_y**2)

mse_log_y = np.mean((pred_log_y - test['total_log'])**2)
mse_log_log = np.mean((pred_log_log_back - test['total'])**2)
print("Mean squared error linjär regression:    ", mse)
print("Mean squared error log transformerad x:  ", mse_log_x)
print("Mean squared error log transformerad y:  ", mse_exp_y)

print("Mean squared error log-log transformerad:", mse_log_log)
```

### residual variansen

```python
# Residualvarians för test data
k = 2
n = len(test)

# Beräkna residualvarians
residual_var = np.sum((test['total'] - pred)**2) / (n - k )
residual_var_y = np.sum((test['total'] - np.exp(pred_log_y))**2) / (n - k )
residual_var_x = np.sum((test['total'] - pred_log)**2) / (n - k )
residual_var_log_log = np.sum((test['total'] - pred_log_log_back)**2) / (n - k )

# Räkna ut R^2
var_y = np.var(test['total'])
r2_linjär = 1 - residual_var/ var_y
r2_exp_y = 1 - residual_var_y / var_y
r2_exp_x = 1 - residual_var_x / var_y
r2_log_log = 1 - residual_var_log_log / var_y


# ----- Printa residualvarians ----- 
print("Var(R)")
print("Residualvarians för log-log modell:         ", residual_var_log_log)
print("Residualvarians för linjär regression:      ", residual_var)
print("Residualvarians för exponentiell modell i y:", residual_var_y)
print("Residualvarians för exponentiell modell i x:", residual_var_x)

# -----  Printa R^2 ----- 
print("\nR^2:")
print("R^2 för log-log modell:         ", r2_log_log)
print("R^2 för linjär regression:      ", r2_linjär)
print("R^2 för exponentiell modell i y:", r2_exp_y)
print("R^2 för exponentiell modell i x:", r2_exp_x)
```
